```python
import numpy as np
import matplotlib.pyplot as plt

# Parametreler
pop_size = 5      # Popülasyon boyutu
dimensions = 1    # Boyut (burada tek değişkenli)
mutation_factor = 0.8
crossover_rate = 0.7
iterations = 10   # İterasyon sayısı

# Hedef fonksiyon: x^2
def objective_function(x):
    return x[0] ** 2

# Başlangıç popülasyonu oluştur
population = np.random.uniform(-10, 10, (pop_size, dimensions))

# Her iterasyonda işlemleri ekrana yazdır ve grafiği çizdir
for i in range(iterations):
    print(f"\nIteration {i + 1}")
    new_population = np.copy(population)
    print(f"Populasyon: {new_population}")
    print("\n")

    # Bireyleri işleyin
    for j in range(pop_size):
        # Üç rastgele farklı birey seç
        indices = np.random.choice(pop_size, 3, replace=False)
        while j in indices:
            indices = np.random.choice(pop_size, 3, replace=False)
        print("İşlem yapılan birey ", population[j])
        a, b, c = population[indices]
        mutant = a + mutation_factor * (b - c)
        print("Rastgele seçilen 3 birey : xr1", a ," xr2",b,"  xr3", c )
        # Mutasyon ve çaprazlama
        print("Fark Vektörü : ",mutant)

        trial = np.copy(population[j])
        rn = np.random.rand()
        print("Random Sayı : ", rn, "  CR : ", crossover_rate)
        if rn < crossover_rate:
            trial[0] = mutant[0]
        ui=trial[0]
        print("ui : ", ui)
        # Değerlendirme
        if objective_function(trial) < objective_function(population[j]):
            new_population[j] = trial
            print("ui seçildi")
        else:
            print("Xi Devam")

        # Hesaplamaları yazdır
        print(f"Birey  {j + 1}: Xi = {population[j][0]:.4f}, ui = {trial[0]:.4f}, "
              f" xi Fitness= {objective_function(population[j]):.4f}, "
              f" ui Fitness = {objective_function(trial):.4f}")
        print("\n\n")

    # Popülasyonu güncelle
    population = new_population

    # Her iterasyon sonunda popülasyonu grafik üzerinde göster
    x_values = np.linspace(-10, 10, 100)
    y_values = x_values**2

    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label='Objective Function (x^2)', color='orange')
    plt.scatter(population, [objective_function(x) for x in population], color='blue', label='Bireyler')
    plt.title(f'Iteration {i + 1}')
    plt.xlabel('Değer')
    plt.ylabel('Fitness (Objective Function)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Sonuç
best_individual = min(population, key=objective_function)
print("\nBest solution found:", best_individual[0])
print("Objective value:", objective_function(best_individual))
```
