```python
import numpy as np
import matplotlib.pyplot as plt

# Parametreler
num_karincalar = 5  # Karınca sayısı
num_sehirler = 10  # Şehir sayısı
iterasyon_sayisi = 10  # İterasyon sayısı
alpha = 1  # Feromon etkisi katsayısı
beta = 2  # Sezgisel bilgi katsayısı
rho = 0.5  # Feromon buharlaşma oranı

# Rastgele şehir konumları (2D düzlem)
np.random.seed(42)
sehirler = np.random.rand(num_sehirler, 2) * 100

# Mesafe matrisi
def hesapla_mesafeler(sehirler):
    mesafeler = np.zeros((len(sehirler), len(sehirler)))
    for i in range(len(sehirler)):
        for j in range(len(sehirler)):
            mesafeler[i][j] = np.linalg.norm(sehirler[i] - sehirler[j])
    return mesafeler

mesafeler = hesapla_mesafeler(sehirler)
print("\n Mesafe Matrisi:")
print(np.round(mesafeler, 4))

# Feromon matrisi
feromon = np.ones((num_sehirler, num_sehirler))

# Çözüm
en_iyi_yol = None
en_iyi_uzunluk = float('inf')

for iterasyon in range(iterasyon_sayisi):
    print(f"\n--- İterasyon {iterasyon + 1} ---")
    yollar = []
    uzunluklar = []

    fig, axs = plt.subplots(1, num_karincalar, figsize=(5 * num_karincalar, 5))
    if num_karincalar == 1:
        axs = [axs]  # Tek karınca durumunda subplot dizisi oluştur

    for karinca, ax in zip(range(num_karincalar), axs):
        # Karıncanın yolunu oluştur
        #np.random.seed(None)
        sehir_sirasi = [np.random.randint(num_sehirler)]
        while len(sehir_sirasi) < num_sehirler:
            mevcut = sehir_sirasi[-1]
            olasilik = [
                (feromon[mevcut][j] ** alpha) * ((1 / mesafeler[mevcut][j]) ** beta)
                if j not in sehir_sirasi else 0
                for j in range(num_sehirler)
            ]
            olasilik = np.array(olasilik) / sum(olasilik)
            sehir_sirasi.append(np.random.choice(range(num_sehirler), p=olasilik))

        yollar.append(sehir_sirasi)

        # Yol uzunluğunu hesapla
        uzunluk = sum(mesafeler[sehir_sirasi[i]][sehir_sirasi[i+1]] for i in range(num_sehirler - 1))
        uzunluk += mesafeler[sehir_sirasi[-1]][sehir_sirasi[0]]  # Döngüyü kapat
        uzunluklar.append(uzunluk)
        print(f"Karınca {karinca + 1}: Yol = {sehir_sirasi}, Uzunluk = {uzunluk:.2f}")

        # Karıncanın yolunu çiz
        yol_x = [sehirler[i][0] for i in sehir_sirasi] + [sehirler[sehir_sirasi[0]][0]]
        yol_y = [sehirler[i][1] for i in sehir_sirasi] + [sehirler[sehir_sirasi[0]][1]]
        ax.plot(yol_x, yol_y, 'o--', alpha=0.7, label=f"Karınca {karinca + 1}")
        ax.scatter(sehirler[:, 0], sehirler[:, 1], c='blue', s=50)
        ax.set_title(f"Karınca {karinca + 1} | Yol Uzunluğu: {uzunluk:.2f}")
        ax.legend()

        # En iyi çözümü güncelle
        if uzunluk < en_iyi_uzunluk:
            en_iyi_uzunluk = uzunluk
            en_iyi_yol = sehir_sirasi

    # Ekranda göster
    plt.tight_layout()
    plt.show()
    print("\nFeromon Matrisi :")
    print(np.round(feromon, 4))
    # Feromon güncellemesi
    feromon *= (1 - rho)  # Buharlaşma
    for yol, uzunluk in zip(yollar, uzunluklar):
        for i in range(num_sehirler - 1):
            print('Yol:', yol[i], ' -->  ', yol[i+1], '=  Feromon:', feromon[yol[i]][yol[i+1]], 'Uzunluk:', uzunluk)

            feromon[yol[i]][yol[i+1]] += 1 / uzunluk
        feromon[yol[-1]][yol[0]] += 1 / uzunluk  # Döngüyü kapat

    print("\nFeromon Matrisi (Güncellenmiş):")
    print(np.round(feromon, 4))

print("\nEn iyi yol:", en_iyi_yol)
print("En iyi uzunluk:", en_iyi_uzunluk)
```
