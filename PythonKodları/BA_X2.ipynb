```python
import numpy as np
import matplotlib.pyplot as plt

# Fonksiyon tanımı
def objective_function(x):
    return x**2

# Yarasa algoritması parametreleri
num_bats = 5
max_iterations = 50
frequency_min = 0
frequency_max = 2
alpha = 0.9  # Yankı gücünün azalma faktörü
gamma = 0.9  # Yankı yayılımının artış faktörü

# Başlangıç noktalarının ve hızların oluşturulması
np.random.seed(42)  # Rastgelelik için sabit değer
pos = np.random.uniform(-10, 10, num_bats)
vel = np.zeros(num_bats)
frequencies = np.zeros(num_bats)
loudness = np.ones(num_bats)  # Yankı gücü (A)
rate = np.random.uniform(0, 1, num_bats)  # Yankı yayılma oranı (r)

# Global en iyi çözümü başlat
best_pos = pos[np.argmin(objective_function(pos))]
best_fitness = objective_function(best_pos)

# Grafik çizimi fonksiyonu
def plot_bats(iteration, pos, obj_func):
    x = np.linspace(-10, 10, 500)
    y = obj_func(x)
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='$f(x) = x^2$', color='blue')
    plt.scatter(pos, objective_function(pos), color='red', label='Bats')
    plt.scatter([best_pos], [best_fitness], color='green', marker='*', s=100, label='Best')
    plt.title(f"Iteration {iteration}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid()
    plt.show()

# Algoritmanın ana döngüsü
for iteration in range(1, max_iterations + 1):
    for i in range(num_bats):
        # Frekans güncellemesi
        frequencies[i] = frequency_min + (frequency_max - frequency_min) * np.random.random()

        # Hız ve konum güncellemesi
        vel[i] = vel[i] + (pos[i] - best_pos) * frequencies[i]
        pos[i] = pos[i] + vel[i]

        # Rastgele yerel arama
        if np.random.random() > rate[i]:
            pos[i] = best_pos + 0.1 * np.random.normal()

        # Sınır kontrolü
        pos[i] = np.clip(pos[i], -10, 10)

        # Yeni çözüm ve güncelleme
        fitness = objective_function(pos[i])
        if fitness < best_fitness and np.random.random() < loudness[i]:
            best_pos = pos[i]
            best_fitness = fitness

            # Yankı parametrelerinin güncellenmesi
            loudness[i] *= alpha
            rate[i] = rate[i] * (1 - np.exp(-gamma * iteration))

    # Her iterasyon sonrası grafiği çiz
    plot_bats(iteration, pos, objective_function)

# En iyi çözümü yazdır
print(f"Best solution: x = {best_pos}, f(x) = {best_fitness}")
```
