```python
import numpy as np
import matplotlib.pyplot as plt

# Sabitler

T = 100  # T parametresi
b = 2    # b parametresi

# t değerlerini oluştur

t = np.linspace(0,100)  # t, 0'dan 200'e kadar eşit aralıklarla

# w(t) fonksiyonunu tanımla
w_t = np.exp(-(b * t / T) ** b)

# Grafiği çiz
plt.figure(figsize=(8, 6))
plt.plot(t, w_t, label=r'$w(t) = e^{-\left(\frac{bt}{T}\right)^b}$', color='blue', lw=2)
plt.title(r' $w(t) = e^{-\left(\frac{bt}{T}\right)^b}$ grafiği', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('w(t)', fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
